# Réduction de la liste des usages

On produit une liste fermée des (67) usages et une analyse fréquentielle de ceux-ci.

Attention, on ne réduit que l'ensemble des usages apparaissant dans les listes d'usages. Les usages figurant dans les 3 colonnes d'usages principal et secondaires n'y sont pas ajoutés. Cela permet notamment de vérifier plus bas que tous les termes d'usage principal ou secondaire figurent bien dans le lexique constitué par fermeture sur l'ensemble des listes d'usages.

La conclusion de l'analyse est qu'aucun des termes qui apparait dans l'une des 3 colonnes de ventilation des usages n'est un terme isolé ne figurant dans aucune des listes d'usages.

# Vérifications pour s'assurer de bases solides

On vérifie par sécurité que chacun des 3 usages principaux font bien partie de la liste.

On vérifie même leur position dans la liste.

20 cas se manifestent où au moins l'un des 3 usages principaux ne figure pas dans la liste.

```python
from use_types_analysis import tricky_split, check_positions

# most of the main uses are first in the list of uses, but this is not systematic : check it
print('LargestPropertyUseType is the first of ListOfAllPropertyUseTypes :',
    (uses_split[0] == use_data.LargestPropertyUseType).all())

# for the rest, we reuse the tricky_split
uses_as_list = use_data.ListOfAllPropertyUseTypes.apply(tricky_split)

main_use_pos = check_positions(use_data, use_data.LargestPropertyUseType, uses_as_list, 'main')
second_use_pos = check_positions(use_data, use_data.SecondLargestPropertyUseType, uses_as_list, 'second')
third_use_pos = check_positions(use_data, use_data.ThirdLargestPropertyUseType, uses_as_list, 'third')
```

Pour compléter cette vérification, listons les usages principaux et secondaires qui ne font pas partie de la liste réduite des usages.

On vérifie bien qu'il n'y a aucun *électron libre*.

```python
from use_types_analysis import not_listed_uses
uses_set = set(unique_uses.index)
print('main uses not listed : ', not_listed_uses(use_data.LargestPropertyUseType, uses_set))
print('second uses not listed : ', not_listed_uses(use_data.SecondLargestPropertyUseType, uses_set))
print('third uses not listed : ', not_listed_uses(use_data.ThirdLargestPropertyUseType, uses_set))
```
